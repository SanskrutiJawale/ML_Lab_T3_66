```python
# Get dataset
df_hear = pd.read_csv('hearing_test.csv')
df_hear.head()
```

```python
# Describe data
df_hear.describe()
```

```python
# Relationship between Salary and Experience
plt.scatter(df_hear['age'], df_hear['physical_score'], color = 'lightcoral')
plt.title('physical_score vs age')
plt.xlabel('age')
plt.ylabel('physical_score')
plt.box(False)
plt.show()
```

```python
# Splitting variables
X = df_hear.iloc[:, :1]  # independent
y = df_hear.iloc[:, 1:]  # dependent
```

```python
# Splitting dataset into test/train
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Regressor model
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Prediction result
y_pred_test = regressor.predict(X_test)     # predicted value of y_test
y_pred_train = regressor.predict(X_train)   # predicted value of y_train
```

```python
# Regressor coefficients and intercept
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```

```python
#Create a model and fit it
model = LinearRegression().fit(X, y)
```

```python
#Get results
r_sq = model.score(X, y)
print(f"coefficient of determination: {r_sq}")


print(f"intercept: {model.intercept_}")


print(f"coefficients: {model.coef_}")
```
